```python
# переменные
a = 5
xmax = 13
xmin = -1
ymax = 13
ymin = -1
shift_x = 8
shift_y = 4
alpha = 45
########################
```

```python
# матрица 5х2 с парами координат x,y вершин квадрата АхА
square = matrix([[1, 1], [1, a], [a, a], [a, 1], [1, 1]])
```

```python
# alpha в радинах, матрица поворота на alpha градусов 2х2
radalpha = math.radians(alpha)
mrot = matrix([[cos(radalpha),sin(radalpha)],[-sin(radalpha),cos(radalpha)]])

# перемножаем 2 матрицы и получаем матрицу повернутого квадрата
square_rotated =  square * mrot
```

```python
# вектор сдвига, матрица сдвига для квадрата
point = vector((shift_x, shift_y))
matrix_shift = matrix([point, point, point, point, point])

# добавляем матрицу сдвига к к матрице перевернутого квадрата и получаем ответ
square_res = square_rotated + matrix_shift
```

```python
# рисуем график 
show(line(square) + line(square_res)
     , xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax
     , aspect_ratio=1, figsize=[4,5])
```

### Непрерывность. Наличие точек разрыва и их классификация:

```python
k1 = limit(y(x), 'right', x=0)
show("Предел справа для точки x1: не существует на D")
k2 = limit(y(x), 'left', x=0)
show("Предел слева для точки x1: ",k2)
k3 = limit(y(x), 'right', x=1)
show("Предел справа для точки x2: ", k3)
k4 = limit(y(x), 'left', x=1)
show("Предел слева для точки x2: не существует на D")
if (abs(k1) == infinity or abs(k2) == infinity):
    show("x1 = 0 - точка разрыва 2 рода")
else:
    show("x1 = 0 - точка разрыва 1 рода")
if (abs(k3) == infinity or abs(k4) == infinity):
    show("x2 = 1 - точка разрыва 2 рода")
else:
    show("x2 = 1 - точка разрыва 1 рода")
```
