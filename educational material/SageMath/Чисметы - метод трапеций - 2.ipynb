```python
dx = 0.6

COLOR_RECT = (1, 0.01, 0.235)
```

```python
# rectangular integration
import numpy as np
p1 = plot(f)
r1 = 0
rects = []
for i, x_0 in enumerate(np.arange(a, b , dx)):
    dy = f(x_0)
    curr = dx * dy
    r1 += curr
    txt = text(r"$i={}, curr={}, result={}$".format(i, str(curr)[:6], str(r1)[:6]), (-4.2,0.5), fontsize=12, color="black")
    
    rects.append(
        polygon2d([(x_0, 0),
                   (x_0 + dx, 0),
                   (x_0 + dx, dy),
                   (x_0, dy)], fill=False, rgbcolor=COLOR_RECT)
    )

    p1+= plot(rects[-1])
    (p1 + txt).show(xmin=0, xmax=7, ymin=-1, ymax=10, aspect_ratio=1)
   
    
r1
```

```python
# plot rectangular integration

p1 = plot(sum(rects)) + plot(f)
p1.show(xmin=0, xmax=7, ymin=-1, ymax=10, aspect_ratio=1)
```

```python
# integration

p1 = plot(f)

r2 = 0
trapeze = []
for i, x_0 in enumerate(np.arange(a, b, dx)):
    dy1 = f(x_0)
    dy2 = f(x_0 + dx)
    curr = dx * ((dy1 + dy2) / 2)
    r2 += curr
    txt = text(r"$i={}, curr={}, result={}$".format(i, str(curr)[:6], str(r2)[:6]), (-4.3,0.5), fontsize=12, color="black")
    
    trapeze.append(
        polygon2d([(x_0, 0),
                   (x_0 + dx, 0),
                   (x_0 + dx, dy2),
                   (x_0, dy1)], fill=False, rgbcolor=COLOR_RECT)
    )
    
    p1+= plot(trapeze[-1])
    (p1 + txt).show(xmin=0, xmax=7, ymin=-1, ymax=10, aspect_ratio=1)
    
    
r2
```
